```python
#Removing white space 
df_columns= df.columns.str.replace(' ', '', regex= True)
#df_columns= df.columns.str.strip()
#df_columns= df.columns.str.replace(r'\s+','', regex =True)

df.columns = df_columns
```

```python
df.rename(columns={'Segment':'segment'}, inplace= True)
df.rename(columns={'Country':'country'}, inplace= True)
df.rename(columns={'DiscountBand':'discount_band'}, inplace= True)
df.rename(columns={'UnitsSold':'units_sold'}, inplace= True)
df.rename(columns={'ManufacturingPrice':'manufacturing_price'}, inplace= True)
df.rename(columns={'SalePrice':'sale_price'}, inplace= True)
df.rename(columns={'GrossSales':'gross_sales'}, inplace= True)
df.rename(columns={'Discounts':'discounts'}, inplace= True)
df.rename(columns={'Sales':'sales'}, inplace= True)
df.rename(columns={'COGS':'cogs'}, inplace= True)
df.rename(columns={'Profit':'profit'}, inplace= True)
df.rename(columns={'Date':'date'}, inplace= True)
```

```python
df
```

```python
# reshaping the dataset
df= df.loc[0:699, ['segment','country','Product','discount_band','units_sold','manufacturing_price','sale_price','gross_sales','discounts','sales','cogs','profit','date']]
```

```python
df
```

```python
remove_dollar_columns =['units_sold', 'manufacturing_price', 'sale_price', 'gross_sales', 'discounts', 'sales', 'cogs', 'profit']
#Remove $ and - from price colmun
for col in remove_dollar_columns:
    if df[col].dtype == 'object':
        df[col]= df[col].str.replace('$','').str.replace('-','').str.replace(',','').str.strip()
```

```python
df.head()
```

```python
# Change following columns 1st to float than to int
df = df.astype({'units_sold': float, 'manufacturing_price': float, 'sale_price': float, 'gross_sales': float,'sales': float, 'cogs': float})
# Change following columns  now to INT
df = df.astype({'units_sold': int, 'manufacturing_price': int, 'sale_price': int, 'gross_sales': int,'sales': int, 'cogs': int})
```

```python
# Checking the information on the data set 
df.info()
```

```python
#add zero in empty columns
df = df.replace({'': '0'}, regex=True)
df.head()
```

```python
#remove () from Profit Column
df['profit'] = df['profit'].str.strip('()')
```

```python
df
```

```python
#convert the Data frame to CSV file
df.to_csv('Financials.csv', index = False)
```
